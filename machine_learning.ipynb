1. Подготовка базовой модели

Обучите простую модель классификации с помощью DecisionTreeClassifier на данных из датасета vehicles_dataset_prepared.csv. Для этого сделайте шаги:

1. Обучите модель дерева решений с зафиксированным random_state на тренировочной выборке.
2. Сделайте предикт на тестовой выборке.
3. Замерьте точность на тестовой выборке и выведите матрицу ошибок. 
4. Удалите фичи с нулевыми весами по feature_importance из тренировочной и тестовой выборок.
5. Заново обучите модель и измерьте качество.

````python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import re
from datetime import datetime

# Загрузка данных
file_path = './vehicles_dataset_old.csv'
print(f"Загрузка файла: {file_path}")

try:
    df = pd.read_csv(file_path)
    print(f"Успешно загружен датасет с {len(df)} samples")
except FileNotFoundError:
    print(f"Файл {file_path} не найден!")
    exit()
except Exception as e:
    print(f"Ошибка при загрузке файла: {e}")
    exit()

print("\nИнформация о датасете:")
print(f"Размер: {df.shape}")
print(f"Колонки: {list(df.columns)}")

# Создаем признаки согласно описанию задания
print("\n" + "="*50)
print("СОЗДАНИЕ ПРИЗНАКОВ ИЗ ИСХОДНЫХ ДАННЫХ")
print("="*50)

df_prepared = df.copy()

# Заполняем пропущенные значения
df_prepared['manufacturer'] = df_prepared['manufacturer'].fillna('unknown')
df_prepared['model'] = df_prepared['model'].fillna('unknown')
df_prepared['fuel'] = df_prepared['fuel'].fillna('unknown')
df_prepared['title_status'] = df_prepared['title_status'].fillna('unknown')
df_prepared['transmission'] = df_prepared['transmission'].fillna('unknown')
df_prepared['state'] = df_prepared['state'].fillna('unknown')
df_prepared['description'] = df_prepared['description'].fillna('')

# Заполняем числовые пропуски
df_prepared['year'] = df_prepared['year'].fillna(df_prepared['year'].median())
df_prepared['odometer'] = df_prepared['odometer'].fillna(df_prepared['odometer'].median())
df_prepared['lat'] = df_prepared['lat'].fillna(df_prepared['lat'].median())
df_prepared['long'] = df_prepared['long'].fillna(df_prepared['long'].median())

# 1. Признаки производителя
df_prepared['is_manufacturer_name'] = (df_prepared['manufacturer'] != 'unknown').astype(int)

# 2. Регион (one-hot encoding для топ-5 регионов)
top_regions = df_prepared['region'].value_counts().head(5).index
for region in top_regions:
    col_name = f"region_{re.sub('[^a-zA-Z0-9]', '_', str(region))}"
    df_prepared[col_name] = (df_prepared['region'] == region).astype(int)

# 3. Тип топлива (one-hot encoding)
fuel_types = df_prepared['fuel'].value_counts().head(3).index
for i, fuel in enumerate(fuel_types, 1):
    df_prepared[f'x0_{i}'] = (df_prepared['fuel'] == fuel).astype(int)

# 4. Производитель (one-hot encoding для топ-5)
top_manufacturers = df_prepared['manufacturer'].value_counts().head(5).index
for i, manuf in enumerate(top_manufacturers, 1):
    df_prepared[f'manufacturer_{i}'] = (df_prepared['manufacturer'] == manuf).astype(int)

# 5. Модель (создаем короткое название)
def get_short_model(model):
    if pd.isna(model) or model == 'unknown':
        return 'unknown'
    # Берем первое слово из названия модели
    model_str = str(model)
    return model_str.split()[0] if ' ' in model_str else model_str

df_prepared['short_model'] = df_prepared['model'].apply(get_short_model)
top_models = df_prepared['short_model'].value_counts().head(3).index
for i, model in enumerate(top_models, 1):
    df_prepared[f'short_model_{i}'] = (df_prepared['short_model'] == model).astype(int)

# 6. Статус (one-hot encoding)
title_statuses = df_prepared['title_status'].value_counts().head(2).index
for i, status in enumerate(title_statuses, 1):
    df_prepared[f'title_status_{i}'] = (df_prepared['title_status'] == status).astype(int)

# 7. Коробка передач
transmissions = df_prepared['transmission'].value_counts().head(2).index
for i, trans in enumerate(transmissions, 1):
    df_prepared[f'transmission_{i}'] = (df_prepared['transmission'] == trans).astype(int)

# 8. Штат (one-hot encoding для топ-5)
top_states = df_prepared['state'].value_counts().head(5).index
for i, state in enumerate(top_states, 1):
    df_prepared[f'state_{i}'] = (df_prepared['state'] == state).astype(int)

# 9. Возрастная категория
current_year = datetime.now().year
df_prepared['car_age'] = current_year - df_prepared['year']
df_prepared['age_category_1'] = (df_prepared['car_age'] <= 5).astype(int)
df_prepared['age_category_2'] = ((df_prepared['car_age'] > 5) & (df_prepared['car_age'] <= 10)).astype(int)
df_prepared['age_category_3'] = (df_prepared['car_age'] > 10).astype(int)

# Инициализируем scaler
scaler = StandardScaler()

# 10. Стандартизированные числовые признаки

# Пробег
df_prepared['std_scaled_odometer'] = scaler.fit_transform(df_prepared[['odometer']])

# Год выпуска
df_prepared['year_std'] = scaler.fit_transform(df_prepared[['year']])

# Широта и долгота
df_prepared['lat_std'] = scaler.fit_transform(df_prepared[['lat']])
df_prepared['long_std'] = scaler.fit_transform(df_prepared[['long']])

# Отношение пробега к цене (избегаем деления на 0)
df_prepared['odometer_price_ratio'] = df_prepared['odometer'] / (df_prepared['price'] + 1)
df_prepared['odometer/price_std'] = scaler.fit_transform(df_prepared[['odometer_price_ratio']])

# Длина описания
df_prepared['desc_len'] = df_prepared['description'].str.len()
df_prepared['desc_len_std'] = scaler.fit_transform(df_prepared[['desc_len']])

# Количество упоминаний модели в описании
def count_model_in_desc(row):
    model_str = str(row['model'])
    desc = str(row['description'])
    if model_str == 'unknown' or not desc:
        return 0
    return desc.lower().count(model_str.lower())

df_prepared['model_in_desc'] = df_prepared.apply(count_model_in_desc, axis=1)
df_prepared['model_in_desc_std'] = scaler.fit_transform(df_prepared[['model_in_desc']])

# Длина названия модели
df_prepared['model_len'] = df_prepared['model'].str.len()
df_prepared['model_len_std'] = scaler.fit_transform(df_prepared[['model_len']])

# Количество слов в названии модели
df_prepared['model_word_count'] = df_prepared['model'].str.split().str.len()
df_prepared['model_word_count_std'] = scaler.fit_transform(df_prepared[['model_word_count']])

# Обработка даты - более безопасный подход
try:
    # Пробуем разные форматы дат
    df_prepared['posting_date'] = pd.to_datetime(df_prepared['posting_date'], errors='coerce', utc=True)
    
    # Если есть пропуски после преобразования, создаем случайные даты
    if df_prepared['posting_date'].isna().any():
        print("Некоторые даты не преобразованы, заполняем случайными...")
        min_date = pd.Timestamp('2020-01-01', tz='UTC')
        max_date = pd.Timestamp('2023-12-31', tz='UTC')
        random_dates = pd.to_datetime(np.random.randint(
            min_date.value, max_date.value, size=df_prepared['posting_date'].isna().sum()
        ), utc=True)
        df_prepared.loc[df_prepared['posting_date'].isna(), 'posting_date'] = random_dates
    
    # Месяц и день недели размещения объявления
    df_prepared['month'] = df_prepared['posting_date'].dt.month
    df_prepared['dayofweek'] = df_prepared['posting_date'].dt.dayofweek
    df_prepared['month_std'] = scaler.fit_transform(df_prepared[['month']])
    df_prepared['dayofweek_std'] = scaler.fit_transform(df_prepared[['dayofweek']])
    
    # Разница в годах между выпуском и размещением объявления
    df_prepared['posting_year'] = df_prepared['posting_date'].dt.year
    df_prepared['diff_years'] = df_prepared['posting_year'] - df_prepared['year']
    df_prepared['diff_years_std'] = scaler.fit_transform(df_prepared[['diff_years']])
    
except Exception as e:
    print(f"Ошибка при обработке дат: {e}")
    # Создаем фиктивные временные признаки если не удалось обработать даты
    df_prepared['month_std'] = np.random.normal(0, 1, len(df_prepared))
    df_prepared['dayofweek_std'] = np.random.normal(0, 1, len(df_prepared))
    df_prepared['diff_years_std'] = np.random.normal(0, 1, len(df_prepared))

# Удаляем исходные столбцы, оставляем только созданные числовые признаки
# Исключаем столбцы с URL и другими строками
feature_columns = [col for col in df_prepared.columns if any(keyword in col for keyword in 
                   ['is_manufacturer', 'region_', 'x0_', 'manufacturer_', 'short_model_', 
                    'title_status_', 'transmission_', 'state_', 'age_category_', 'std_scaled', 
                    'year_std', 'lat_std', 'long_std', 'odometer/price_std', 'desc_len_std', 
                    'model_in_desc_std', 'model_len_std', 'model_word_count_std', 'month_std', 
                    'dayofweek_std', 'diff_years_std'])]

# Убедимся, что все признаки числовые
numeric_columns = []
for col in feature_columns:
    if col in df_prepared.columns:
        # Проверяем, что столбец числовой
        if pd.api.types.is_numeric_dtype(df_prepared[col]):
            numeric_columns.append(col)
        else:
            print(f"Удаляем нечисловой столбец: {col}")

print(f"Создано числовых признаков: {len(numeric_columns)}")
print("Примеры созданных признаков:")
print(numeric_columns[:20])

# Проверяем, что есть признаки для обучения
if len(numeric_columns) == 0:
    print("Ошибка: не создано ни одного числового признака!")
    exit()

# Подготовка данных для модели
x = df_prepared[numeric_columns]
y = df_prepared['price_category']

print(f"\nФинальный размер данных для обучения:")
print(f"Признаки: {x.shape}")
print(f"Целевая переменная: {y.shape}")
print(f"Классы в целевой переменной: {y.unique()}")

# Проверяем типы данных
print(f"\nТипы данных признаков:")
print(x.dtypes.value_counts())

# Разделение на тренировочную и тестовую выборки
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.3, random_state=42, stratify=y
)

print(f"\nРазмеры выборок:")
print(f"Обучающая выборка: {x_train.shape}")
print(f"Тестовая выборка: {x_test.shape}")

# Шаг 1: Обучение базовой модели дерева решений
print("\n" + "="*50)
print("ШАГ 1: БАЗОВАЯ МОДЕЛЬ")
print("="*50)

base_model = DecisionTreeClassifier(random_state=42, max_depth=10)
base_model.fit(x_train, y_train)

# Шаг 2: Предсказание на тестовой выборке
y_pred_base = base_model.predict(x_test)

# Шаг 3: Оценка точности и матрица ошибок
accuracy_base = accuracy_score(y_test, y_pred_base)
print(f"Точность базовой модели: {accuracy_base:.4f}")

print("\nМатрица ошибок базовой модели:")
cm_base = confusion_matrix(y_test, y_pred_base)
print(cm_base)

# Визуализация матрицы ошибок
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.heatmap(cm_base, annot=True, fmt='d', cmap='Blues', 
            xticklabels=y.unique(), yticklabels=y.unique())
plt.title('Матрица ошибок\nБазовая модель')
plt.ylabel('Фактические значения')
plt.xlabel('Предсказанные значения')

# Анализ важности признаков
print("\nТоп-10 самых важных признаков в базовой модели:")
feature_importance = pd.DataFrame({
    'feature': x_train.columns,
    'importance': base_model.feature_importances_
}).sort_values('importance', ascending=False)

print(feature_importance.head(10))

# Визуализация важности признаков
plt.subplot(1, 2, 2)
top_features = feature_importance.head(10)
sns.barplot(data=top_features, x='importance', y='feature')
plt.title('Топ-10 важных признаков\nБазовая модель')
plt.tight_layout()
plt.show()

# Шаг 4: Удаление фич с нулевой важностью
print("\n" + "="*50)
print("ШАГ 4: УДАЛЕНИЕ ПРИЗНАКОВ С НУЛЕВОЙ ВАЖНОСТЬЮ")
print("="*50)

zero_importance_features = feature_importance[feature_importance['importance'] == 0]['feature'].tolist()
print(f"Признаки с нулевой важностью ({len(zero_importance_features)}):")
if zero_importance_features:
    print(zero_importance_features[:10])
    if len(zero_importance_features) > 10:
        print(f"... и еще {len(zero_importance_features) - 10} признаков")
else:
    print("Нет признаков с нулевой важностью!")

# Удаляем признаки с нулевой важностью
if zero_importance_features:
    x_train_reduced = x_train.drop(zero_importance_features, axis=1)
    x_test_reduced = x_test.drop(zero_importance_features, axis=1)
else:
    x_train_reduced = x_train.copy()
    x_test_reduced = x_test.copy()

print(f"\nРазмеры выборок после удаления признаков:")
print(f"Обучающая выборка: {x_train_reduced.shape}")
print(f"Тестовая выборка: {x_test_reduced.shape}")

# Шаг 5: Обучение новой модели на уменьшенном наборе признаков
print("\n" + "="*50)
print("ШАГ 5: УЛУЧШЕННАЯ МОДЕЛЬ")
print("="*50)

improved_model = DecisionTreeClassifier(random_state=42, max_depth=10)
improved_model.fit(x_train_reduced, y_train)

# Предсказание на тестовой выборке
y_pred_improved = improved_model.predict(x_test_reduced)

# Оценка точности и матрица ошибок
accuracy_improved = accuracy_score(y_test, y_pred_improved)
print(f"Точность улучшенной модели: {accuracy_improved:.4f}")

print("\nМатрица ошибок улучшенной модели:")
cm_improved = confusion_matrix(y_test, y_pred_improved)
print(cm_improved)

# Визуализация сравнения моделей
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.heatmap(cm_improved, annot=True, fmt='d', cmap='Greens',
           xticklabels=y.unique(), yticklabels=y.unique())
plt.title('Матрица ошибок\nУлучшенная модель')
plt.ylabel('Фактические значения')
plt.xlabel('Предсказанные значения')

# Важность признаков в улучшенной модели
feature_importance_improved = pd.DataFrame({
    'feature': x_train_reduced.columns,
    'importance': improved_model.feature_importances_
}).sort_values('importance', ascending=False)

plt.subplot(1, 2, 2)
top_features_improved = feature_importance_improved.head(10)
sns.barplot(data=top_features_improved, x='importance', y='feature')
plt.title('Топ-10 важных признаков\nУлучшенная модель')
plt.tight_layout()
plt.show()

# Сравнение результатов
print("\n" + "="*50)
print("СРАВНЕНИЕ РЕЗУЛЬТАТОВ")
print("="*50)
print(f"Точность базовой модели: {accuracy_base:.4f}")
print(f"Точность улучшенной модели: {accuracy_improved:.4f}")
print(f"Разница в точности: {accuracy_improved - accuracy_base:+.4f}")

print(f"\nКоличество признаков:")
print(f"Исходное: {x_train.shape[1]}")
print(f"После удаления: {x_train_reduced.shape[1]}")
print(f"Удалено признаков: {len(zero_importance_features)}")

if x_train.shape[1] > 0:
    efficiency = x_train_reduced.shape[1] / x_train.shape[1] * 100
    print(f"Эффективность: {efficiency:.1f}% от исходного количества")

# Анализ результатов
if accuracy_improved >= accuracy_base:
    print("\n✅ Улучшенная модель показывает одинаковую или лучшую точность с меньшим количеством признаков!")
else:
    print("\n⚠️  Точность немного снизилась, но модель стала проще и быстрее.")

print("\nТоп-5 самых важных признаков в улучшенной модели:")
print(feature_importance_improved.head(5))
````

````python
# Ваш код здесь
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

print("="*70)
print("АНАЛИЗ ВЛИЯНИЯ ФИЧЕЙ НА МОДЕЛЬ")
print("="*70)

# Загрузка исходного датасета
print("Загрузка датасета vehicles_dataset_old.csv...")
df_old = pd.read_csv('./vehicles_dataset_old.csv')
print(f"Размер датасета: {df_old.shape}")

# Шаг 1: Подготовка сырых данных
print("\n" + "="*50)
print("ШАГ 1: ПОДГОТОВКА СЫРЫХ ДАННЫХ")
print("="*50)

# Создаем копию датасета
df_raw = df_old.copy()

# Оставляем только числовые колонки (исключаем все строковые)
numeric_columns = ['id', 'year', 'odometer', 'lat', 'long']
target_column = 'price_category'

# Проверяем, что все числовые колонки существуют
available_numeric_columns = [col for col in numeric_columns if col in df_raw.columns]
print(f"Доступные числовые колонки: {available_numeric_columns}")

# Создаем X из числовых колонок
X_raw = df_raw[available_numeric_columns]
y_raw = df_raw[target_column]

print(f"Размеры сырых данных:")
print(f"Признаки: {X_raw.shape}")
print(f"Целевая переменная: {y_raw.shape}")
print(f"Колонки: {list(X_raw.columns)}")

# Проверяем пропущенные значения
print(f"\nПропущенные значения в сырых данных:")
print(X_raw.isnull().sum())

# Заполняем пропущенные значения медианой
X_raw_filled = X_raw.fillna(X_raw.median())

# Разделение на тренировочную и тестовую выборки
X_train_raw, X_test_raw, y_train_raw, y_test_raw = train_test_split(
    X_raw_filled, y_raw, test_size=0.3, random_state=42, stratify=y_raw
)

print(f"\nРазмеры выборок для сырых данных:")
print(f"Обучающая выборка: {X_train_raw.shape}")
print(f"Тестовая выборка: {X_test_raw.shape}")

# Шаг 2: Обучение лучшей модели на сырых данных
print("\n" + "="*50)
print("ШАГ 2: ОБУЧЕНИЕ НА СЫРЫХ ДАННЫХ")
print("="*50)

# Используем лучшие параметры
best_params = {
    'n_estimators': 400,
    'max_depth': 25,
    'min_samples_split': 5,
    'min_samples_leaf': 2,
    'max_features': 'sqrt',
    'bootstrap': True,
    'random_state': 50
}

print("Лучшие параметры модели:")
for param, value in best_params.items():
    print(f"  {param}: {value}")

# Обучаем модель на сырых данных
rf_raw = RandomForestClassifier(**best_params)
rf_raw.fit(X_train_raw, y_train_raw)

# Предсказание и оценка
y_pred_raw = rf_raw.predict(X_test_raw)
accuracy_raw = accuracy_score(y_test_raw, y_pred_raw)

print(f"\nРезультаты на сырых данных:")
print(f"Точность: {accuracy_raw:.4f}")

# Шаг 3: Создание и оценка предобработанных данных для сравнения
print("\n" + "="*50)
print("ШАГ 3: СОЗДАНИЕ ПРЕДОБРАБОТАННЫХ ДАННЫХ ДЛЯ СРАВНЕНИЯ")
print("="*50)

# Создаем engineered features как в предыдущих заданиях, но ТОЛЬКО числовые
df_processed = df_old.copy()

# Заполняем пропущенные значения
df_processed['manufacturer'] = df_processed['manufacturer'].fillna('unknown')
df_processed['model'] = df_processed['model'].fillna('unknown')
df_processed['fuel'] = df_processed['fuel'].fillna('unknown')
df_processed['title_status'] = df_processed['title_status'].fillna('unknown')
df_processed['transmission'] = df_processed['transmission'].fillna('unknown')
df_processed['state'] = df_processed['state'].fillna('unknown')
df_processed['description'] = df_processed['description'].fillna('')

# Заполняем числовые пропуски
df_processed['year'] = df_processed['year'].fillna(df_processed['year'].median())
df_processed['odometer'] = df_processed['odometer'].fillna(df_processed['odometer'].median())
df_processed['lat'] = df_processed['lat'].fillna(df_processed['lat'].median())
df_processed['long'] = df_processed['long'].fillna(df_processed['long'].median())

# Создаем основные engineered features (ТОЛЬКО числовые)
scaler = StandardScaler()

# 1. Стандартизированные числовые признаки
df_processed['year_std'] = scaler.fit_transform(df_processed[['year']])
df_processed['odometer_std'] = scaler.fit_transform(df_processed[['odometer']])
df_processed['lat_std'] = scaler.fit_transform(df_processed[['lat']])
df_processed['long_std'] = scaler.fit_transform(df_processed[['long']])

# 2. Отношение пробега к цене
df_processed['odometer_price_ratio'] = df_processed['odometer'] / (df_processed['price'] + 1)
df_processed['odometer_price_std'] = scaler.fit_transform(df_processed[['odometer_price_ratio']])

# 3. Возраст автомобиля
current_year = 2023  # Примерный год из данных
df_processed['car_age'] = current_year - df_processed['year']
df_processed['car_age_std'] = scaler.fit_transform(df_processed[['car_age']])

# 4. One-hot encoding для топ-3 производителей (создает числовые колонки)
top_manufacturers = df_processed['manufacturer'].value_counts().head(3).index
for i, manuf in enumerate(top_manufacturers, 1):
    df_processed[f'manufacturer_{i}'] = (df_processed['manufacturer'] == manuf).astype(int)

# 5. One-hot encoding для топ-3 регионов (создает числовые колонки)
top_regions = df_processed['region'].value_counts().head(3).index
for i, region in enumerate(top_regions, 1):
    col_name = f"region_{i}"
    df_processed[col_name] = (df_processed['region'] == region).astype(int)

# 6. Топливо (one-hot encoding) - создает числовые колонки
fuel_types = df_processed['fuel'].value_counts().head(2).index
for i, fuel in enumerate(fuel_types, 1):
    df_processed[f'fuel_{i}'] = (df_processed['fuel'] == fuel).astype(int)

# 7. Длина описания
df_processed['desc_len'] = df_processed['description'].str.len()
df_processed['desc_len_std'] = scaler.fit_transform(df_processed[['desc_len']])

# 8. Длина названия модели
df_processed['model_len'] = df_processed['model'].str.len()
df_processed['model_len_std'] = scaler.fit_transform(df_processed[['model_len']])

# Выбираем ТОЛЬКО числовые engineered features
engineered_columns = [col for col in df_processed.columns if any(keyword in col for keyword in 
                      ['_std', 'manufacturer_', 'region_', 'fuel_', 'car_age', 'odometer_price', 'desc_len', 'model_len'])]

# Убедимся, что все колонки числовые
numeric_engineered_columns = []
for col in engineered_columns:
    if pd.api.types.is_numeric_dtype(df_processed[col]):
        numeric_engineered_columns.append(col)
    else:
        print(f"Удаляем нечисловую колонку: {col}")

print(f"Создано числовых engineered features: {len(numeric_engineered_columns)}")
print("Примеры engineered features:")
print(numeric_engineered_columns[:15])

# Подготовка engineered данных
X_engineered = df_processed[numeric_engineered_columns]
y_engineered = df_processed[target_column]

# Удаляем колонки с NaN значениями
X_engineered = X_engineered.fillna(0)

print(f"\nРазмеры engineered данных:")
print(f"Признаки: {X_engineered.shape}")

# Разделение engineered данных
X_train_eng, X_test_eng, y_train_eng, y_test_eng = train_test_split(
    X_engineered, y_engineered, test_size=0.3, random_state=42, stratify=y_engineered
)

print(f"Размеры выборок для engineered данных:")
print(f"Обучающая выборка: {X_train_eng.shape}")
print(f"Тестовая выборка: {X_test_eng.shape}")

# Обучаем модель на engineered данных
rf_engineered = RandomForestClassifier(**best_params)
rf_engineered.fit(X_train_eng, y_train_eng)

y_pred_eng = rf_engineered.predict(X_test_eng)
accuracy_engineered = accuracy_score(y_test_eng, y_pred_eng)

print(f"\nРезультаты на engineered данных:")
print(f"Точность: {accuracy_engineered:.4f}")

# Сравнение точности
print("\n" + "="*50)
print("СРАВНЕНИЕ ТОЧНОСТИ")
print("="*50)

print("Сравнение точности:")
print(f"Сырые данные: {accuracy_raw:.4f}")
print(f"Engineered данные: {accuracy_engineered:.4f}")
print(f"Разница: {accuracy_engineered - accuracy_raw:+.4f}")

improvement = ((accuracy_engineered - accuracy_raw) / accuracy_raw) * 100
print(f"Улучшение за счет feature engineering: {improvement:+.2f}%")

# Визуализация сравнения
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
models = ['Сырые данные', 'Engineered данные']
accuracies = [accuracy_raw, accuracy_engineered]
colors = ['red', 'green']

bars = plt.bar(models, accuracies, color=colors, alpha=0.7)
plt.axhline(y=0.76, color='blue', linestyle='--', linewidth=2, label='Цель: 0.76')
plt.ylabel('Точность')
plt.title('Сравнение: сырые vs engineered данные')
plt.ylim(0, 1)

# Добавляем значения на столбцы
for bar, accuracy in zip(bars, accuracies):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, 
             f'{accuracy:.4f}', ha='center', va='bottom')

plt.legend()

# Матрица ошибок для сырых данных
plt.subplot(1, 2, 2)
cm_raw = confusion_matrix(y_test_raw, y_pred_raw)
sns.heatmap(cm_raw, annot=True, fmt='d', cmap='Reds',
            xticklabels=y_raw.unique(), yticklabels=y_raw.unique())
plt.title('Матрица ошибок\n(сырые данные)')
plt.ylabel('Фактические значения')
plt.xlabel('Предсказанные значения')

plt.tight_layout()
plt.show()

# Шаг 4: Анализ важности признаков
print("\n" + "="*50)
print("ШАГ 4: АНАЛИЗ ВАЖНОСТИ ПРИЗНАКОВ")
print("="*50)

# Важность признаков в сырых данных
feature_importance_raw = pd.DataFrame({
    'feature': X_train_raw.columns,
    'importance': rf_raw.feature_importances_
}).sort_values('importance', ascending=False)

print("Важность признаков в сырых данных:")
print(feature_importance_raw)

# Важность признаков в engineered данных
feature_importance_eng = pd.DataFrame({
    'feature': X_train_eng.columns,
    'importance': rf_engineered.feature_importances_
}).sort_values('importance', ascending=False)

print("\nТоп-10 самых важных признаков в engineered данных:")
print(feature_importance_eng.head(10))

# Визуализация важности признаков
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.barplot(data=feature_importance_raw, x='importance', y='feature', palette='viridis')
plt.title('Важность признаков (сырые данные)')
plt.xlabel('Важность')

plt.subplot(1, 2, 2)
top_eng_features = feature_importance_eng.head(10)
sns.barplot(data=top_eng_features, x='importance', y='feature', palette='coolwarm')
plt.title('Топ-10 важных признаков (engineered данные)')
plt.xlabel('Важность')

plt.tight_layout()
plt.show()

# Шаг 5: Поиск 3х самых важных фичей
print("\n" + "="*50)
print("ШАГ 5: ПОИСК 3Х САМЫХ ВАЖНЫХ ФИЧЕЙ")
print("="*50)

# Выбираем 3 самых важных признака из engineered данных
top_3_features = feature_importance_eng.head(3)
print(f"🎯 ТРИ САМЫХ ВАЖНЫХ ФИЧИ:")
for i, (_, row) in enumerate(top_3_features.iterrows(), 1):
    print(f"{i}. {row['feature']}: {row['importance']:.4f}")

# Анализ что представляют собой эти фичи
feature_explanations = {
    'year_std': 'Стандартизированный год выпуска автомобиля',
    'odometer_std': 'Стандартизированный пробег (odometer)',
    'odometer_price_std': 'Стандартизированное отношение пробега к цене',
    'car_age_std': 'Стандартизированный возраст автомобиля',
    'lat_std': 'Стандартизированная широта',
    'long_std': 'Стандартизированная долгота',
    'desc_len_std': 'Стандартизированная длина описания',
    'model_len_std': 'Стандартизированная длина названия модели',
    'manufacturer_1': 'Признак принадлежности к топ-1 производителю',
    'manufacturer_2': 'Признак принадлежности к топ-2 производителю',
    'manufacturer_3': 'Признак принадлежности к топ-3 производителю',
    'region_1': 'Признак принадлежности к топ-1 региону',
    'region_2': 'Признак принадлежности к топ-2 региону',
    'region_3': 'Признак принадлежности к топ-3 региону',
    'fuel_1': 'Признак основного типа топлива',
    'fuel_2': 'Признак второго типа топлива'
}

print("\n📊 АНАЛИЗ ТОП-3 ФИЧЕЙ:")
for feature in top_3_features['feature']:
    explanation = feature_explanations.get(feature, 'Важный engineered признак')
    print(f"  • {feature}: {explanation}")

# Шаг 6: Анализ влияния фичей на качество
print("\n" + "="*50)
print("ШАГ 6: АНАЛИЗ ВЛИЯНИЯ ФИЧЕЙ НА КАЧЕСТВО")
print("="*50)

print("🔍 КЛЮЧЕВЫЕ ВЫВОДЫ:")
print(f"1. Feature engineering улучшил точность на {improvement:+.2f}%")
print(f"2. С {accuracy_raw:.4f} (сырые) до {accuracy_engineered:.4f} (engineered)")
print(f"3. Количество признаков: {X_train_raw.shape[1]} → {X_train_eng.shape[1]}")
print(f"4. Улучшение за счет: стандартизации, one-hot encoding, создания новых признаков")

print(f"\n🎯 ТОП-3 САМЫХ ВЛИЯТЕЛЬНЫХ ФИЧИ:")
for i, (_, row) in enumerate(top_3_features.iterrows(), 1):
    feature_name = row['feature']
    importance = row['importance']
    explanation = feature_explanations.get(feature_name, 'Важный engineered признак')
    print(f"   {i}. {feature_name} ({importance:.4f})")
    print(f"      {explanation}")

# Дополнительный анализ: что было упущено в сырых данных
print("\n" + "="*50)
print("ЧТО БЫЛО УПУЩЕНО В СЫРЫХ ДАННЫХ:")
print("="*50)

print("1. 📅 Временные характеристики:")
print("   - Возраст автомобиля (car_age)")
print("   - Стандартизированный год выпуска")

print("2. 🔢 Отношения между переменными:")
print("   - Отношение пробега к цене")
print("   - Стандартизированные значения")

print("3. 🏷️ Категориальные переменные:")
print("   - Производители (one-hot encoding)")
print("   - Регионы (one-hot encoding)") 
print("   - Типы топлива (one-hot encoding)")

print("4. 📏 Текстовые метаданные:")
print("   - Длина описания автомобиля")
print("   - Длина названия модели")

print("\n" + "="*70)
print("ФИНАЛЬНЫЕ ВЫВОДЫ")
print("="*70)

print("✅ Feature engineering КРИТИЧЕСКИ ВАЖЕН для качества моделей")
print(f"✅ Улучшение точности: {improvement:+.2f}%")
print("✅ Самые важные фичи обычно: стандартизированные числовые признаки и отношения")
print("✅ One-hot encoding категориальных переменных значительно помогает")
print(f"✅ Цель 0.76 {'ДОСТИГНУТА' if accuracy_engineered > 0.76 else 'НЕ ДОСТИГНУТА'}")

if accuracy_engineered > 0.76:
    print("🎉 ПОЗДРАВЛЯЕМ! Модель преодолела целевой порог 0.76!")
else:
    print("💡 Для достижения цели 0.76 нужно:")
    print("   - Больше feature engineering")
    print("   - Настройка гиперпараметров")
    print("   - Добавление дополнительных признаков")
````
